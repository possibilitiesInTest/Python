```python
import seaborn as sns
import numpy as np
import pandas as pd

%matplotlib inline
```

```python
sns.set_style('darkgrid')
```

```python
penguins = sns.load_dataset('penguins')
```

```python
penguins.head()
```

```python
sns.displot(penguins['body_mass_g'], bins=5, kde=True, rug=True)
```

```python
sns.kdeplot(penguins['body_mass_g'])
```

```python
sns.rugplot(penguins['body_mass_g'])
```

```python
sns.jointplot(x='bill_length_mm', y='bill_depth_mm', data=penguins, kind='hex')
```

```python
sns.pairplot(penguins, hue="island")
```
