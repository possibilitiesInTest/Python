```python
import seaborn as sns
import numpy as np
import pandas as pd

%matplotlib inline
```

```python
sns.set_style('darkgrid')
```

```python
penguins = sns.load_dataset('penguins')
```

```python
penguins.head()
```

```python
sns.barplot(x="sex", y="body_mass_g", data=penguins, estimator=np.std, hue='island')
```

```python
sns.countplot(x='sex', data=penguins, hue='species')
```

```python
sns.boxplot(x='species', y='flipper_length_mm', data=penguins, hue='sex')
```

```python
sns.violinplot(x='island', y='flipper_length_mm', data=penguins, hue='sex', split=True)
```

```python
sns.stripplot(x='island', y='flipper_length_mm', data=penguins, hue='sex', dodge=True, jitter=False)
```

```python
sns.swarmplot(x='island', y='flipper_length_mm', data=penguins, hue='island')
```

```python
sns.catplot(x='island', y='flipper_length_mm', data=penguins, kind='strip', hue='island')
```
