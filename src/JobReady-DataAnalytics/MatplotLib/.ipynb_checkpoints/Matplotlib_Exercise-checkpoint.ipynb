```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# import data using the statement below
```

```python
seAsia = pd.read_csv('WHO-COVID-19-SEAsia.csv')
```

```python
seAsia['Country'].unique()
```

```python
# create a new DataFrame 'india' from 'seAsia' where the country is India
# plot x=Date_reported with y=Cumulative_cases where the country is India
```

```python
india = seAsia[seAsia["Country"]=="India"]
india
```

```python
plt.plot(india['Date_reported'],india['Cumulative_cases'])
```

```python
# plot x=Date_reported with y=New_cases where the country is India using the Object Oriented Method
# and a figure size of 15 by 5
```

```python
fig = plt.figure(figsize=(15,5))

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes.plot(india['Date_reported'], india['New_cases'])
```

```python
# create a new DataFrame 'bangla' from 'seAsia' where the country is Bangladesh
# plot two subplots (2 rows and 1 column of plots)
#  - top plot should have x=Date_reported and y=New_cases for india
#  - bottom plot should have x=Date_reported and y=New_cases for bangla
#  - each plot should have a title and labels for x-axis and y-axis
```

```python
bangla = seAsia[seAsia['Country']=='Bangladesh']
bangla
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(15,10))

axes[0].plot(india['Date_reported'], india['New_cases'])
axes[0].set_title('Daily New Cases for India')
axes[0].set_xlabel('Date Reported')
axes[0].set_ylabel('New Cases')


axes[1].plot(bangla['Date_reported'], bangla['New_cases'])
axes[1].set_title('Daily New Cases for Bangladesh')
axes[1].set_xlabel('Date Reported')
axes[1].set_ylabel('New Cases')
```

```python
# combine the two plots into a single plot with a legend to differentiate the line for each country
```

```python
fig = plt.figure(figsize=(15,5))

axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(india['Date_reported'], india['New_cases'], label='India')
axes.plot(bangla['Date_reported'], bangla['New_cases'], label='Bangladesh')

axes.set_title('Daily New Cases for India and Bangladesh')
axes.set_xlabel('Date Reported')
axes.set_ylabel('New Cases')
axes.legend()
```
