```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
x = np.linspace(0,10,21)
```

```python
y = x ** 2
```

```python
fig = plt.figure(figsize=(15,5), dpi=100)

axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
```

```python
fig = plt.figure(figsize=(5,3))

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='orange', lw=2, alpha=0.8, ls="-", marker="v", markersize="15", markerfacecolor="yellow", markeredgecolor="yellow", markeredgewidth=3)
```

```python
fig = plt.figure(figsize=(15,5))
ax =fig.add_axes([0.1,0.1,0.8,0.8])
ax.plot(x,y,color='orange',lw=3,ls="-")

ax.set_xlim([0,4])
ax.set_ylim([0,40])
```

```python
fig.savefig("fig2.png", dpi=200)
```
