```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
x = np.linspace(0,10,21)
```

```python
x
```

```python
y = x ** 2
```

```python
y
```

```python
plt.plot(x,y)
# plt.show()
```

```python
plt.hist(x)
```

```python
plt.scatter(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'b')
plt.subplot(1,2,2)
plt.plot(y,x,'g')
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.4,0.4])
axes1.plot(x,y)
axes1.set_title('Left_Plot')

axes2 = fig.add_axes([0.6,0.1,0.4,0.4])
axes2.plot(y,x)
axes2.set_title('Right_Plot')


```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.9,0.9])
axes1.plot(x,y)
axes1.set_title('Main Plot')

axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.4])
axes2.plot(y,x)
axes2.set_title('Insert Plot')
```
