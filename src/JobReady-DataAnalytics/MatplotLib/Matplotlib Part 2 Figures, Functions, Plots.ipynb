```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
x = np.linspace(0,10,21)
```

```python
y = x ** 2
```

```python
fig = plt.figure(figsize=(15,5), dpi=100)

axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(15,5), dpi=100)

axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()
```

```python
fig=plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, x+2, label = "x + 2")
ax.plot(x, x**2, label = "x ** 2")

ax.legend(loc=(0.2,0.8))
```

```python
fig.savefig('my_fig.jpg', dpi=200)
```
